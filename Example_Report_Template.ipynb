```python
import pandas as pd
import numpy as np
```

```python
all_teams =  pd.read_csv("2022_Project_distance_Matrix.csv")
```

## Data Cleaning
In this section provide code for converting the raw data into clean and usable data structures (if needed)

```python
index = []

for ind,row in all_teams.iterrows():
    for i in range(len(row)):
        if row[i] == 0:
            index.append(all_teams.columns[i])
all_teams["team"] = index
all_teams = all_teams.set_index("team")
all_teams
```

## Data Modeling

This section that builds a model.  This doesn't have to be all that advanced. You should probably start with some simple statistical models such as an average distance matrix.

```python
avg_df = pd.DataFrame(dtype = float)
team_names = all_teams.columns
for team in team_names:
    ind_team = all_teams.loc[all_teams.index == team ]
    ind_team = ind_team.apply(np.mean,axis = 0)
    avg_df = pd.concat([avg_df, ind_team],axis = 1)
avg_df.columns = all_teams.columns
avg_df

    
```
