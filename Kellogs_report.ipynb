## Data input
Import the csv files.

```python
import pandas as pd
pd.set_option("display.max_rows", None)
data = pd.read_csv("2022_Project_distance_Matrix.csv")
data
```

## Data Modeling
We created a distance matrix based on the csv file.

```python
matri = pd.DataFrame([data[data[i]==0].mean().values.tolist() for i in data.columns])
matri.columns = data.columns
matri.index = data.columns
matri = matri.round(2)
matri
```

```python
matri.describe()
```

## Data Visualization
This is the heapmap of the distance matrix. Dark means the similar group, Bright means they think they are totally different project.

It read row by row, 
for example, the first row is the ARFL's opinion on other teams' project.


```python
import numpy as np
import seaborn as sns
sns.set(rc = {'figure.figsize':(10,8)})
ax = sns.heatmap(matri)
```

```python
sns.set(rc = {'figure.figsize':(8,6)})
ax2 = sns.clustermap(matri)
```
